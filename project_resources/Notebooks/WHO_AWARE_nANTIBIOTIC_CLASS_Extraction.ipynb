## Data Validation and Update Process
Let's validate the ATC codes and add WHO AWARE Classification and Antimicrobial Class information.

```python
# Load the completed antimicrobials data
print("Loading antimicrobials data...")
df_completed = pd.read_excel('C:\\Personal_Projects\\Astro\\project_resources\\Completed_Data_Tables_withVersions\\Antimicrobials_with_ATC_COMPLETE.xlsx')
print(f"Loaded {len(df_completed)} antimicrobials")

# Clean up WHO EML data - skip the description rows and reset index
df_who_eml_clean = df_who_eml_ALL.iloc[2:].copy()  # Skip first two rows
df_who_eml_clean.columns = ['Antibiotic', 'Class', 'ATC_code', 'Category', 'Listed_on_EML', 'Unnamed_5', 'Unnamed_6', 'Unnamed_7']
df_who_eml_clean = df_who_eml_clean.reset_index(drop=True)

print("\nWHO EML data shape:", df_who_eml_clean.shape)
print("\nSample of WHO EML data:")
display(df_who_eml_clean.head())
```

```python
# Create mapping dictionaries from WHO EML data
atc_to_class = dict(zip(df_who_eml_clean['ATC_code'], df_who_eml_clean['Class']))
atc_to_category = dict(zip(df_who_eml_clean['ATC_code'], df_who_eml_clean['Category']))
name_to_class = dict(zip(df_who_eml_clean['Antibiotic'].str.lower(), df_who_eml_clean['Class']))
name_to_category = dict(zip(df_who_eml_clean['Antibiotic'].str.lower(), df_who_eml_clean['Category']))

# Function to find best match for antibiotic name
def find_matching_antibiotic(name):
    name_lower = name.lower()
    # Try exact match first
    if name_lower in name_to_class:
        return name_to_class[name_lower], name_to_category[name_lower]
    
    # Try removing common modifiers
    base_name = name_lower.split('/')[0].strip()
    if base_name in name_to_class:
        return name_to_class[base_name], name_to_category[base_name]
    
    return None, None

# Update the dataframe
df_updated = df_completed.copy()

# Initialize counters
updated_count = 0
missing_count = 0
validation_issues = []

# Process each row
for idx, row in df_updated.iterrows():
    atc_code = row['ATC_CODE']
    antibiotic_name = row['ANTIMICROBIAL']
    
    # Try to get class and category by ATC code first
    antimicrobial_class = atc_to_class.get(atc_code)
    aware_category = atc_to_category.get(atc_code)
    
    # If not found by ATC code, try by name
    if not antimicrobial_class or not aware_category:
        class_by_name, category_by_name = find_matching_antibiotic(antibiotic_name)
        antimicrobial_class = antimicrobial_class or class_by_name
        aware_category = aware_category or category_by_name
    
    # Update the row
    if antimicrobial_class or aware_category:
        if pd.isna(df_updated.at[idx, 'ANTIMICROBIAL_CLASS']):
            df_updated.at[idx, 'ANTIMICROBIAL_CLASS'] = antimicrobial_class
        if pd.isna(df_updated.at[idx, 'WHO_AWARE_CLASSIFICATION']):
            df_updated.at[idx, 'WHO_AWARE_CLASSIFICATION'] = aware_category
        updated_count += 1
    else:
        missing_count += 1
        validation_issues.append(f"No match found for: {antibiotic_name} (ATC: {atc_code})")

# Print summary
print("\n=== Update Summary ===")
print(f"Total antimicrobials processed: {len(df_updated)}")
print(f"Successfully updated: {updated_count}")
print(f"Missing information: {missing_count}")

# Show sample of updated data
print("\n=== Sample of Updated Data ===")
display(df_updated.head())

# Show validation issues
if validation_issues:
    print("\n=== Validation Issues ===")
    for issue in validation_issues[:10]:
        print(issue)
    if len(validation_issues) > 10:
        print(f"...and {len(validation_issues) - 10} more issues")

# Save updated data
output_file = 'C:\\Personal_Projects\\Astro\\project_resources\\Completed_Data_Tables_withVersions\\Antimicrobials_with_ATC_COMPLETE_AWARE.xlsx'
df_updated.to_excel(output_file, index=False)
print(f"\nUpdated data saved to: {output_file}")

# Create summary of AWARE classifications
print("\n=== AWARE Classification Distribution ===")
print(df_updated['WHO_AWARE_CLASSIFICATION'].value_counts())

print("\n=== Antimicrobial Classes Distribution ===")
print(df_updated['ANTIMICROBIAL_CLASS'].value_counts().head(10))
print(f"...and {len(df_updated['ANTIMICROBIAL_CLASS'].value_counts()) - 10} more classes")
```

```python
# Perform data quality checks
print("\n=== Data Quality Checks ===")

# Check for missing values
print("\nMissing values in each column:")
print(df_updated.isna().sum())

# Check unique values in WHO_AWARE_CLASSIFICATION
print("\nUnique WHO AWARE Classifications:")
print(df_updated['WHO_AWARE_CLASSIFICATION'].unique())

# Verify ATC code format
def is_valid_atc(code):
    if pd.isna(code):
        return False
    return bool(re.match(r'^[A-Z][0-9]{2}[A-Z]{2}[0-9]{2}$', str(code)))

invalid_atc = df_updated[~df_updated['ATC_CODE'].apply(lambda x: is_valid_atc(x) if pd.notna(x) else True)]
if not invalid_atc.empty:
    print("\nInvalid ATC codes found:")
    print(invalid_atc[['ANTIMICROBIAL', 'ATC_CODE']])

# Compare with WHO EML data
print("\nMatching statistics with WHO EML data:")
matched_atc = df_updated['ATC_CODE'].isin(df_who_eml_clean['ATC_code'])
print(f"ATC codes matched with WHO EML: {matched_atc.sum()} out of {len(df_updated)}")

# Save validation report
validation_report = f"""Validation Report ({pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')})

Total antimicrobials: {len(df_updated)}
Successfully matched: {updated_count}
Missing information: {missing_count}

Missing values by column:
{df_updated.isna().sum()}

WHO AWARE Classification distribution:
{df_updated['WHO_AWARE_CLASSIFICATION'].value_counts()}

Validation issues:
{chr(10).join(validation_issues)}
"""

validation_file = 'C:\\Personal_Projects\\Astro\\project_resources\\Completed_Data_Tables_withVersions\\validation_report.txt'
with open(validation_file, 'w') as f:
    f.write(validation_report)

print(f"\nDetailed validation report saved to: {validation_file}")
```
